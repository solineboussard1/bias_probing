```python
import pandas as pd

# Read in the CSV files
complete_exploded = pd.read_csv('complete_exploded.csv')
topic_mapping = pd.read_csv('topic_mapping_anxiety_management.csv')
data_w_pca = pd.read_csv('data_with_pca.csv')
# Merge the dataframes
# Since topic_mapping has multiple rows per prompt, we'll merge on Prompt to get all combinations
merged_df = pd.merge(complete_exploded, topic_mapping, on='Prompt', how='left')
merged_df = pd.merge(merged_df, data_w_pca, on='Prompt', how='left')
```

```python
import numpy as np
from scipy.optimize import linear_sum_assignment
import matplotlib.pyplot as plt

# Function to calculate agreement score
def calculate_agreement(cluster_labels, topic_labels, pca_cluster_labels):
    n_clusters = len(np.unique(cluster_labels))
    n_topics = len(np.unique(topic_labels))
    n_pca_clusters = len(np.unique(pca_cluster_labels))

    # Create confusion matrices
    confusion_matrix_cluster_topic = np.zeros((n_clusters, n_topics))
    confusion_matrix_cluster_pca = np.zeros((n_clusters, n_pca_clusters))
    confusion_matrix_topic_pca = np.zeros((n_topics, n_pca_clusters))

    for i in range(len(cluster_labels)):
        confusion_matrix_cluster_topic[cluster_labels[i], topic_labels[i]] += 1
        confusion_matrix_cluster_pca[cluster_labels[i], pca_cluster_labels[i]] += 1
        confusion_matrix_topic_pca[topic_labels[i], pca_cluster_labels[i]] += 1

    # Use Hungarian algorithm to find optimal matchings
    row_ind_ct, col_ind_ct = linear_sum_assignment(-confusion_matrix_cluster_topic)
    row_ind_cp, col_ind_cp = linear_sum_assignment(-confusion_matrix_cluster_pca)
    row_ind_tp, col_ind_tp = linear_sum_assignment(-confusion_matrix_topic_pca)

    # Calculate agreement scores
    total_matches_ct = confusion_matrix_cluster_topic[row_ind_ct, col_ind_ct].sum()
    total_matches_cp = confusion_matrix_cluster_pca[row_ind_cp, col_ind_cp].sum()
    total_matches_tp = confusion_matrix_topic_pca[row_ind_tp, col_ind_tp].sum()

    agreement_score_ct = total_matches_ct / len(cluster_labels)
    agreement_score_cp = total_matches_cp / len(cluster_labels)
    agreement_score_tp = total_matches_tp / len(cluster_labels)

    return (agreement_score_ct, agreement_score_cp, agreement_score_tp), (row_ind_ct, col_ind_ct), (row_ind_cp, col_ind_cp), (row_ind_tp, col_ind_tp)

# Convert Cluster, Dominant_Topic, and PCA Cluster to numeric labels
merged_df['Cluster'] = pd.Categorical(merged_df['Cluster']).codes
merged_df['Dominant_Topic'] = pd.Categorical(merged_df['Dominant_Topic']).codes
merged_df['pca_cluster_number'] = pd.Categorical(merged_df['pca_cluster_number']).codes

# Calculate agreement scores
agreement_scores, cluster_topic_mapping, cluster_pca_mapping, topic_pca_mapping = calculate_agreement(
    merged_df['Cluster'].values,
    merged_df['Dominant_Topic'].values,
    merged_df['pca_cluster_number'].values
)

print(f"Cluster-Topic Agreement Score: {agreement_scores[0]:.3f}")
print(f"Cluster-PCA Agreement Score: {agreement_scores[1]:.3f}")
print(f"Topic-PCA Agreement Score: {agreement_scores[2]:.3f}")

# Create mapping dictionaries
cluster_to_topic = {cluster: topic for cluster, topic in zip(cluster_topic_mapping[0], cluster_topic_mapping[1])}
cluster_to_pca = {cluster: pca for cluster, pca in zip(cluster_pca_mapping[0], cluster_pca_mapping[1])}
topic_to_pca = {topic: pca for topic, pca in zip(topic_pca_mapping[0], topic_pca_mapping[1])}

# Function to check if a row agrees with the optimal mappings
def check_agreement(row):
    cluster_topic_agree = int(cluster_to_topic[row['Cluster']] == row['Dominant_Topic'])
    cluster_pca_agree = int(cluster_to_pca[row['Cluster']] == row['pca_cluster_number'])
    topic_pca_agree = int(topic_to_pca[row['Dominant_Topic']] == row['pca_cluster_number'])
    return cluster_topic_agree, cluster_pca_agree, topic_pca_agree

# Calculate agreement for each row
merged_df[['cluster_topic_agree', 'cluster_pca_agree', 'topic_pca_agree']] = merged_df.apply(check_agreement, axis=1, result_type='expand')

# Visualize agreement
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 6))

scatter1 = ax1.scatter(merged_df['pca_one'], merged_df['pca_two'],
                       c=merged_df['cluster_topic_agree'], cmap='RdYlGn', alpha=0.6)
ax1.set_title('Cluster-Topic Agreement')
fig.colorbar(scatter1, ax=ax1, label='Agreement')

scatter2 = ax2.scatter(merged_df['pca_one'], merged_df['pca_two'],
                       c=merged_df['cluster_pca_agree'], cmap='RdYlGn', alpha=0.6)
ax2.set_title('Cluster-PCA Agreement')
fig.colorbar(scatter2, ax=ax2, label='Agreement')

scatter3 = ax3.scatter(merged_df['pca_one'], merged_df['pca_two'],
                       c=merged_df['topic_pca_agree'], cmap='RdYlGn', alpha=0.6)
ax3.set_title('Topic-PCA Agreement')
fig.colorbar(scatter3, ax=ax3, label='Agreement')

for ax in (ax1, ax2, ax3):
    ax.set_xlabel('PCA Dimension 1')
    ax.set_ylabel('PCA Dimension 2')

plt.tight_layout()
plt.show()

# Print the optimal mappings
print("\nOptimal Cluster to Topic Mapping:")
for cluster, topic in cluster_to_topic.items():
    print(f"Cluster {cluster} -> Topic {topic}")

print("\nOptimal Cluster to PCA Cluster Mapping:")
for cluster, pca in cluster_to_pca.items():
    print(f"Cluster {cluster} -> PCA Cluster {pca}")

print("\nOptimal Topic to PCA Cluster Mapping:")
for topic, pca in topic_to_pca.items():
    print(f"Topic {topic} -> PCA Cluster {pca}")

```
